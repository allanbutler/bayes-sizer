```python
import numpy as np
import matplotlib.pyplot as plt
from bayessizer.mde import fixed_sample_mde
```

```python
# Define parameters
alpha = 0.05
beta = 0.2
sigma_sq = 1
N_values = np.arange(100, 5000, 100)  # Sample sizes
mu = 1

# Compute MDE for different sample sizes
mde_values = [fixed_sample_mde(alpha, beta, sigma_sq, N, mu) for N in N_values]

# Plot MDE vs Sample Size
plt.figure(figsize=(8, 5))
plt.plot(N_values, mde_values, marker='o', linestyle='-')
plt.xlabel('Sample Size (N)')
plt.ylabel('Minimum Detectable Effect (MDE)')
plt.title('Fixed-Sample MDE vs Sample Size')
plt.grid()
plt.show()

# Display a single example calculation
example_mde = fixed_sample_mde(alpha, beta, sigma_sq, 1000, mu)
print(f"Example MDE for N=1000: {example_mde:.4f}")
```
